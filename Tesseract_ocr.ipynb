```python
import os
import pandas as pd
import shutil
from google.colab import files

uploaded = files.upload()

file_list = []
input_folder_path = "/content/input_ocr"  # Files will be uploaded to the content directory
output_folder_path = "/content/output_ocr"

for filename in uploaded.keys():
    if filename.endswith(".pdf"):
        # Construct the source and destination paths
        source_path = "/content/" + filename  # Files are initially uploaded to /content/
        destination_path = input_folder_path + "/" + filename

        # Move the file
        shutil.move(source_path, destination_path)

        # Add the filename to the file_list
        file_list.append(filename)

print("The following files will be converted:")
pd.DataFrame(file_list, columns=["Filename"])
```

```python
import os
from google.colab import files

# ... (Your existing code for file conversion)
for filename in file_list:
  print("Converting : " + filename)
  input_filename = input_folder_path+'/'+filename
  output_filename = output_folder_path+'/'+filename[:-4]+'_ocr.pdf'
  !ocrmypdf --force-ocr "{input_filename}" "{output_filename}"

  # Download the converted file
  files.download(output_filename)
```

```python
import os
from google.colab import files
import zipfile

# Create a zip file
zip_filename = output_folder_path + "/processed_files.zip"
with zipfile.ZipFile(zip_filename, 'w') as zipf:
    for filename in file_list:
        processed_filename = output_folder_path + "/" + filename[:-4]+'_ocr.pdf'
        zipf.write(processed_filename, arcname=filename)

# Download the zip file
files.download(zip_filename)
```
